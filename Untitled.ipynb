```python
import duckdb

duckdb.sql("""
CREATE OR REPLACE VIEW commune AS
FROM 'https://www.insee.fr/fr/statistiques/fichier/7766585/v_commune_2024.csv';
""")
```

```python
duckdb.sql("""
SELECT *
  FROM commune
""")
```

```python
import pandas as pd

duckdb.sql("""
SELECT *
  FROM commune
""").to_df()
```

```python
duckdb.sql("""
SELECT distinct REG
  FROM commune
  WHERE DEP LIKE '976'
""")
```

```python
duckdb.sql("""
SELECT COUNT(*)
  FROM commune
  WHERE DEP LIKE '973'
""")
```

```python
duckdb.sql("""
SELECT COUNT(*)
  FROM commune
  WHERE DEP IN ('02','60','80')
      AND NCC ILIKE '%L%L%L%L%'
""").to_df()
```

```python
duckdb.sql("""
SELECT COUNT(*),DEP
  FROM commune
  GROUP BY DEP
""").to_df()
```

```python
duckdb.sql("""
SELECT COUNT(1),DEP
  FROM commune
  GROUP BY DEP
  ORDER BY 1 DESC
""").to_df()
```

```python
duckdb.sql("""
SELECT COUNT(DISTINCT COM) AS 'nbr communes',DEP
  FROM commune
  WHERE DEP IS NOT NULL
  GROUP BY DEP
  ORDER BY 1 DESC
  OFFSET 19 ROWS FETCH NEXT 1 ROWS ONLY
""").to_df()
```

```python
duckdb.sql("""
SELECT COUNT(DISTINCT COM) AS 'nbr communes',DEP
  FROM commune
  WHERE DEP IS NOT NULL
  GROUP BY DEP
  ORDER BY 1 DESC
""").to_df()
```
